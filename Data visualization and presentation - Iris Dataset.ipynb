## Import Iris Dataset from Seaborn

```python
iris=sns.load_dataset('iris')
iris.head()
```

## Sumary of Dataset (number of samples, datatypes etc)

```python
iris.describe() #Creates and pandas dataframe with some statistical info
```

```python
print(iris.info())
```

```python
gbys=iris.groupby('species') #Groupby object
```

```python
print(gbys.size())
```

## Data presentation and visualization

```python
sns.pairplot(iris,hue='species', size=5, aspect=1)
```

```python
sns.pairplot(data=iris,hue='species',diag_kind='kde',size=5,aspect=1)
```

### Overall Histogram

```python
iris.hist(edgecolor='black', linewidth=1.2,figsize=(12,8));
plt.show();
```

### Violin Plots

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.violinplot(x='species',y='petal_length',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species',y='petal_width',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species',y='sepal_length',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species',y='sepal_width',data=iris)
```

### Boxplot

```python
iris.boxplot(by='species', figsize=(12,8)) #By using dataframes method
```

```python
#By using seaborn
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.boxplot(x='species',y='petal_length',data=iris)
plt.subplot(2,2,2)
sns.boxplot(x='species',y='petal_width',data=iris)
plt.subplot(2,2,3)
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.subplot(2,2,4)
sns.boxplot(x='species',y='sepal_width',data=iris)
```
